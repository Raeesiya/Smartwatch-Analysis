### correlation

```python
#To find correlation between numerical columns
df.corr()
```

```python
#Visualizing Correlations Between Numerical Features Using a Heatmap
plt.figure(figsize=(20, 8))
sns.heatmap(df.corr(), annot=True, cmap='Spectral')
plt.show()
```


* TotalSteps and TotalDistance have the high correlation.
* TotalSteps and LightActiveDistance have the medium correlation.
* LightlyActiveMinutes and SedentaryMinutes have the low correlation.

### Creating a new column

```python
#Creating a Total Activity Minutes Column by Summing Different Activity Levels
df["Total Minutes"]=df.VeryActiveMinutes+df.FairlyActiveMinutes+df.LightlyActiveMinutes+df.SedentaryMinutes
df["Total Minutes"]
```

# visualization

## Bar Graph

##### Distribution of calories burned by each persons

```python
#ploting the histogram
plt.figure(figsize=(8,6))
plt.hist(df['Calories'],bins=10,color='hotpink',edgecolor='black')
plt.xlabel('Calories Burned')
plt.ylabel('Frequency')
plt.title('Distribution of calories burned by each persons')
plt.show()
```

##### Distribution of total steps per day

```python
#ploting the histogram
plt.figure(figsize=(10, 6))
plt.hist(df['TotalSteps'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Total Steps per Day')
plt.ylabel('Frequency')
plt.title('Distribution of Total Steps per Day')
plt.show()
```

##### Average calories burned by activity level

```python
#ploting the histogram
activity_levels = ['VeryActive','FairlyActive','LightlyActive','Sedentary']
avg_calories_burned = [500, 300, 150, 50] 
plt.figure(figsize=(10,6))
plt.bar(activity_levels, avg_calories_burned, color=['blue','green','orange','red'])
plt.title('Average Calories Burned by Activity Level')
plt.xlabel('Activity Level')
plt.ylabel('Average Calories Burned')
plt.show()
```

### Days of week and counts

```python
#Extracts the day name from the ActivityDate column
df['day_of_week'] = df['ActivityDate'].dt.day_name()
#Returns the count of occurrences for each day of the week 
df['day_of_week'].value_counts()
```

## Pie chart

##### Distribution of dates

```python
plt.figure(figsize=(7,7))
plt.pie(df['day_of_week'].value_counts(), labels=df['day_of_week'].value_counts().index, autopct='%1.1f%%', startangle=140)
plt.title('Dstribution of dates')
plt.show()
```

## Scatter plot

##### Total steps vs Calories

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['TotalSteps'],df['Calories'],color='red',alpha=0.5)
plt.xlabel("TotalSteps")
plt.ylabel("Calories")
plt.title("Scatter Plot of Total Steps vs Calories")
plt.show()
```
